## Analysis functions:
We want to perform 3 analysis: descriptive, detective and predictive:
- Descriptive: 
    - Goals:
        - describe the whole dataset.
        - how many values per variable.
        - map out the geographic location.
        - comparison between different categories.
- Detective: 
    - Goals:
        - Which variables lead to a higher value of "charges" (correlation matrix),
        - which category is affected the most (men, women, smokers, non-smokers)
- Predictive: 
    - Goals:
        - Build a ML model to predict the insurance charges
    

### Descriptive analysis:

Building a countplot for differnet categorical variables:
sex and smoker

```python
#importing seaborn and setting the theme(optional)
import seaborn as sns
sns.set_style("dark")
```

```python
#sex
ax = sns.countplot(data=df_init, x="sex", color="skyblue")
ax.patches[0].set_facecolor('pink')
ax.bar_label(ax.containers[0])
```

```python
#smoker
ax = sns.countplot(data=df_init, x="smoker", color="mediumseagreen")
ax.patches[0].set_facecolor('indianred')
ax.bar_label(ax.containers[0])
```

```python
#region
ax = sns.countplot(data=df_init, x="region", palette="Set3")
ax.bar_label(ax.containers[0])
```

How many smokers per region

```python
ax = sns.countplot(data=df_init, x="region", hue="smoker",palette="Set2")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
```

How many smokers per sex

```python
ax = sns.countplot(data=df_init, x="sex", hue="smoker", palette="Set2")

ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
```

Age distribution across genders

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
ax = sns.countplot(data=df_init, x="age", hue="sex", palette="Set2")
```

BMI distribution

```python
ax = sns.histplot(data=df_init, x="bmi")
```
